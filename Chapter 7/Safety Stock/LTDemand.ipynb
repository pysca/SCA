### Calculating Average Daily Demand

```python
Avg_daily_demand = Sales.groupby('Id').Quantity.mean()    #getting daily demand for each product based on 'Id'
```

```python
Avg_daily_demand = Avg_daily_demand.to_frame().reset_index()  # alternatively, you can use pd.DataFrame()
```

```python
Avg_daily_demand.rename(columns={'Quantity':'DailyDemand'}, inplace=True)  #change column name 
```

```python
Avg_daily_demand
```

```python
df = pd.read_csv('temp3.csv')
df
```

```python
new_df = df.merge(Avg_daily_demand, on='Id')
new_df
```

### Calculating Lead Time Demand
LeadTime Demand = Average Lead Time * Average Demand

```python
new_df['LTDemand'] = new_df.AvgLT * new_df.DailyDemand
new_df.LTDemand
```

```python
ltd = new_df.groupby('Id', sort=False).LTDemand.mean()
```

```python
ltd
```

```python
ltd = ltd.to_frame()
```

```python
ltd
```

```python
plt.hist(ltd.LTDemand,bins=30);
```

```python
#Standard deviation of demand

stdDmd = Sales.groupby('Id').Quantity.std()
stdDmd.isna().sum()
```

```python
StdDmd = stdDmd.to_frame().reset_index()
```

```python
StdDmd.rename(columns={'Quantity':'StdDmd'}, inplace=True)
```

```python
new_df = new_df.merge(StdDmd, on='Id')
new_df
```
